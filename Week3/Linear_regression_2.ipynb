# Cost function
### Let's define a function to calculate the cost.
image.png

```python
def computeCost(X,y, theta):

    m = len(y)
    h = X.dot(theta)
    J = 1/(2*m)*(np.sum((h-y)**2))
    return J
```

### Let's print the cost using current values.



```python
cost = computeCost(X,y, theta)

# Your code to print cost
print(cost)
```

# Gradient Descent algorithm for linear regression.
### Let's implement the Gradient Descent algorithm to reduce cost.

### Note- We already initialized $\theta_{0},$ $\theta_{1}$. What are their initial values?

image.png

```python
def gradientDescent(X, y, theta, alpha, iter):

    J_history = []
    m = len(y)

    for i in range(iter):
        h = X.dot(theta)
        theta = theta - (alpha/m)*(X.T.dot(h-y))
        J_history.append(computeCost(X, y, theta))
    return theta, J_history

```

# Running Gradient descent

### Let's execute the gradient descent process for a specific number of iterations of your choice (please aim for a minimum of 1000), using a fixed learning rate of $\alpha = 0.01$.

### You can select the number of iterations according to your preference, but ensure it exceeds 1000. Importantly, refrain from copying the iterations used by your peers; it should be chosen independently by you.

```python
# Set iteration values of your choice (Keep it more than 1000)
iter = 1400
alpha = 0.01
new_theta, J_history = gradientDescent(X, y, theta, alpha, iter)

# Your code to print new theta values.
print(new_theta)
```

## Splitting the dataset into the Training set and Test set


```python
from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 1/3, random_state = 0)
```

```python
print (X_train.shape)
print (X_test.shape)

print (y_train.shape)
print (y_test.shape)
```

## Training the Simple Linear Regression model on the Training set

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

## Predicting the Test set results


```python
y_pred = regressor.predict(X_test)
```

## Visualising the Training set results


```python
plt.scatter(X_train[:, 1], y_train, color = 'red')
plt.plot(X_train[:, 1], regressor.predict(X_train), color = 'blue')
plt.title('Salary vs Experience (Training set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```
